# Предсказание атака/не атака методом Байеса

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, accuracy_score

%config IPCompleter.greedy=True 
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
from sklearn.externals import joblib
# для сохранения обученных моделей
```

```python
small = pd.read_csv('small', sep = '	')
```

```python
small
```

```python
small = small.drop('Unnamed: 0', axis = 1)
```

```python
small
```

```python
y = small['string.4']
x = small.drop('string.4', axis = 1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, random_state=10, test_size=0.7)
```

```python
grid_bayes = GaussianNB()
```

```python
%%time
grid_bayes.fit(X_train, y_train)
```

```python
accuracy_score(y_test, grid_bayes.predict(X_test))
```
